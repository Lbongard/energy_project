```python
def rf_reg_grid_search(X_train, y_train, param_distributions):
    """
    Performs a grid search for a random forest classifier using cross-validation.

    Parameters
    ----------
    X_train : array-like of shape (n_samples, n_features)
        The training input samples.
    y_train : array-like of shape (n_samples,)
        The target values.
    param_grid : dict
        Dictionary with parameters names (str) as keys and lists of parameter
        settings to try as values.

    Returns
    -------
    best_params : dict
        Dictionary with the best parameters found during the grid search.
    """

    # Define the random forest classifier
    rf = RandomForestRegressor()

    # Perform a grid search using cross-validation
    grid_search = RandomizedSearchCV(rf, 
                                     param_distributions=param_distributions, 
                                     cv=5, 
                                     n_iter=5,
                                     scoring='neg_root_mean_squared_error')
    grid_search.fit(X_train, y_train)

    # Print the best hyperparameters
    print("Best hyperparameters found during grid search:")
    best_params = grid_search.best_params_
    print(best_params)

    return best_params
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, None],
    'min_samples_split': [5, 10, 15],
    'min_samples_leaf': [2, 4, 8]
}
```

```python
def rf_grid_search(X_train, y_train, param_distributions):
    """
    Performs a grid search for a random forest classifier using cross-validation.

    Parameters
    ----------
    X_train : array-like of shape (n_samples, n_features)
        The training input samples.
    y_train : array-like of shape (n_samples,)
        The target values.
    param_grid : dict
        Dictionary with parameters names (str) as keys and lists of parameter
        settings to try as values.

    Returns
    -------
    best_params : dict
        Dictionary with the best parameters found during the grid search.
    """

    # Define the random forest classifier
    rf = RandomForestClassifier()

    # Perform a grid search using cross-validation
    grid_search = RandomizedSearchCV(rf, 
                                     param_distributions=param_distributions, 
                                     cv=5, 
                                     n_iter=5,
                                     scoring='roc_auc')
    grid_search.fit(X_train, y_train)

    # Print the best hyperparameters
    print("Best hyperparameters found during grid search:")
    best_params = grid_search.best_params_
    print(best_params)

    return best_params
```

# LSTM Model - DART Values

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dropout
from tensorflow.keras.preprocessing import sequence
# fix random seed for reproducibility
tf.random.set_seed(42)
```

```python
# For the models in this analysis, we will train using data from 2020-2021 and test against data in 2022
train = df[df.index.year < 2022].dropna(subset=endog + ['future_dart'])
X_train = train[endog]
y_train_contemp = train['DART']
y_train_future = train['future_dart']
y_train_30 = train['DART_spike_30_binary']
y_train_45 = train['DART_spike_45_binary']
y_train_60 = train['DART_spike_60_binary']

test = df[df.index.year == 2022].dropna(subset=endog + ['future_dart'])
X_test = test[endog]
y_test_contemp = test['DART']
y_test_future = test['future_dart']
y_test_30 = test['DART_spike_30_binary']
y_test_45 = test['DART_spike_45_binary']
y_test_60 = test['DART_spike_60_binary']
```

```python
train_val = np.array(df[df.index.year < 2022].dropna(subset=endog + ['future_dart']))
```

```python
# Converting data to np arrays
train_val = df[df.index.year < 2022].dropna(subset=endog + ['future_dart'])
X_train_val = train_val[endog]
y_train_val = train_val['future_dart']

train_len = round(X_train_val.shape[0]*.8)
X_train = np.array(X_train_val[:train_len])
y_train = np.array(y_train_val[:train_len])

X_val = np.array(X_train_val[train_len:])
y_val = np.array(y_train_val[train_len:])

test = df[df.index.year == 2022].dropna(subset=endog + ['future_dart'])
X_test = np.array(test[endog])
y_test = np.array(test['future_dart'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

scaled_X = {
    'train': scaler.transform(X_train),
    'val': scaler.transform(X_val),
    'test': scaler.transform(X_test),
}
```

```python
# Reshape values to be 3D [samples, timesteps, features]
X_train_scaled = scaled_X['train'].reshape((scaled_X['train'].shape[0], 1, scaled_X['train'].shape[1]))
X_val_scaled = scaled_X['val'].reshape((scaled_X['val'].shape[0], 1, scaled_X['val'].shape[1]))
X_test_scaled = scaled_X['test'].reshape((scaled_X['test'].shape[0], 1, scaled_X['test'].shape[1]))
```

```python
print(X_train_scaled.shape, X_val_scaled.shape, X_test_scaled.shape)
```

```python
from tensorflow.keras.optimizers import Adam

model = Sequential()
# model.add(LSTM(15))
model.add(LSTM(100, input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2])))
model.add(Dropout(0.2))
# model.add(LSTM(80))
# model.add(LSTM(50))
model.add(Dense(1))
model.compile(
    optimizer=Adam(),
    loss='mean_squared_error',
    metrics=[
        'mean_absolute_error',
    ]
)
```

```python
model.summary()
```

```python
%%time

from tensorflow.keras.callbacks import TerminateOnNaN, ReduceLROnPlateau, EarlyStopping


history = model.fit(
    X_train_scaled,
    y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=20,
    batch_size=70,
    shuffle=False,
    callbacks=[
        TerminateOnNaN(),
        ReduceLROnPlateau(
            monitor='loss',
            min_delta=0.1,
            patience=5,
            cooldown=10,
            verbose=1
        ),
        EarlyStopping(
            monitor='loss',
            min_delta=0.1,
            patience=5
        )
    ]
)
```

```python
y_hat = model.predict(X_test_scaled)
```
